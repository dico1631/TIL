# requests 할 때 parameter 넘기는 방법

```python
def get_html(url):
    html = ''
    res = requests.get(url)
    if res.status_code == 200:
        res.encoding = 'utf-8'
    html = res.text
    return html
```

```python
import requests

#get방식 외에는 ? 뒤에 쓰는 게 안 통해서 이런 방식을 이용
param = {'age' : 30, 'name' : 'python', 'hobby' : 'painting'}
res = requests.get('http://ggoreb.com/python/request.jsp',params = param)
print(res.status_code)
print(res.text)
```

# beutifulsoup 사용하기
> 'https://finance.naver.com/' 에서 코스피 지수 가져오기

```python
# encoding을 utf-8로 했는데도 깨짐
result = get_html("https://finance.naver.com/")
result
```

```python
def get_html(url):
    html = ''
    res = requests.get(url)
    if res.status_code == 200:
        res.encoding = None  # 깨지는 걸 막기위한 옵션
    html = res.text
    return html
```

```python
result = get_html("https://finance.naver.com/")
result
```

```python
# conda install beautifulsoup4 로 cmd에서 설치
from bs4 import BeautifulSoup
soup = BeautifulSoup(result, 'html.parser')  #항상 html.parser이라고 쓰면 html 다 해결가능
```

```python
# find는 태그만 접근 가능, select 함수는 css 선택자로 접근 가능 / 더 많은 걸 할 수 있음.
nums = soup.select('.num_quot > .num')
for num in nums:
    print(num.getText())
```

```python
import requests

params = {'query' : '서울특별시 동작구 신대방동 395'}
headers = {'Authorization' : 'KakaoAK d4be7b479f4b4cbd99bd19ae87f88b4b'}
res = requests.get('https://dapi.kakao.com/v2/local/search/address.xml', params=params, headers=headers)
res.text
```

```python
from bs4 import BeautifulSoup as bs

soup = bs(res.text, 'html.parser')
addr_list = soup.find_all('address')
for addr in addr_list:
    x = addr.find('x')
    y = addr.find('y')
    name = addr.find('address_name')
    print(x, y, name)
```
